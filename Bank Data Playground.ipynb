```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
bank = pd.read_csv("bank.csv", sep=";")
```

```python
bank
```

```python
bank["y"] = bank["y"].map({'yes': 1, 'no': 0})
bank["default"] = bank["default"].map({'yes': 1, 'no': 0})
bank["housing"] = bank["housing"].map({'yes': 1, 'no': 0})
bank["loan"] = bank["loan"].map({'yes': 1, 'no': 0})
```

```python
bank.describe()
```

```python
bank.info()
```

```python
bank.isnull().sum()
```

```python
bank.corr()
```

```python
bank_dummied = pd.get_dummies(bank)
```

```python
import sklearn
from sklearn.model_selection import train_test_split
```

```python
x = bank_dummied.loc[:, bank_dummied.columns != 'y']
y = bank_dummied[["y"]]
```

```python
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.33, random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
logistic_regression_model = LogisticRegression()
```

```python
logistic_regression_model.fit(X_train, y_train)
```

```python
X_train.shape
```

```python
y_train.shape
```

```python
prediction = logistic_regression_model.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix
```

```python
confusion_matrix(y_test,prediction)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test,prediction)
```

```python
from keras.models import Sequential
from keras.layers import Dense
```

```python
model = Sequential()
model.add(Dense(12, input_dim=48, activation='relu'))
model.add(Dense(8, activation='relu'))
model.add(Dense(8, activation='relu'))
model.add(Dense(8, activation='relu'))
model.add(Dense(8, activation='relu'))
model.add(Dense(4, activation='relu'))
model.add(Dense(8, activation='relu'))
model.add(Dense(8, activation='relu'))
model.add(Dense(8, activation='relu'))
model.add(Dense(8, activation='relu'))
model.add(Dense(4, activation='relu'))
model.add(Dense(1, activation='sigmoid'))
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
model.fit(x, y, epochs=500, batch_size=20)
```

```python
_, accuracy = model.evaluate(x, y)
print('Accuracy: %.2f' % (accuracy*100))
```
